```python
data1_copy[[ 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI']]=data1_copy[[ 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI']].replace(0,np.nan)
```

```python
data1_copy.hist(figsize=(15,15))
```

```python
data1_copy['Glucose'].fillna(data1_copy['Glucose'].mean(),inplace=True)
```

```python
data1_copy.isnull().sum()
```

```python
data1_copy['BloodPressure'].fillna(data1_copy['BloodPressure'].mean(),inplace=True)
```

```python
data1_copy.isnull().sum()
```

```python
data1_copy['SkinThickness'].fillna(data1_copy['SkinThickness'].mean(),inplace=True)
```

```python
data1_copy['Insulin'].fillna(data1_copy['Insulin'].mean(),inplace=True)
```

```python
data1_copy['BMI'].fillna(data1_copy['BMI'].mean(),inplace=True)
```
