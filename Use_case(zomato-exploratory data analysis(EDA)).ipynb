```python
zomato1.rename(columns={'approx_cost(for two people)':'cost','listed_in(type)':'type',
                       'listed_in(city)':'city'},inplace=True)
```

```python
zomato1.columns
```

```python
zomato1['cost']
```

```python
zomato1['cost']=zomato1['cost'].apply(lambda a:a.replace(',','.'))
```

```python
zomato1['cost']=zomato1['cost'].astype('float')
```

```python
zomato1.info()
```

```python
zomato1['rate'].unique()
```

```python
zomato1=zomato1.loc[zomato1['rate']!='NEW']
```

```python
zomato1.shape
```

```python
zomato1=zomato1.loc[zomato1['rate']!='-']
```

```python
zomato1.shape
```

```python
zomato1['rate'].unique()
```

```python
lambda_slash=lambda a:a.replace('/5','') if type(a)==str else a
```

```python
zomato1['rate']=zomato1['rate'].apply(lambda_slash).str.strip().astype('float')
```

```python
zomato1.info()
```

```python
zomato1_copy=zomato1.copy()
```

```python
zomato1['online_order'].replace(('Yes','No'),(True,False),inplace=True)
```

```python
zomato1['online_order']
```

```python
zomato11_copy=zomato1.copy()
```

```python
def convert_numbers(z):
    for c in z.columns[~z.columns.isin(['rate','cost','votes'])]:
                       z[c]=z[c].factorize()[0]
    return z
```

```python
zomato1_encoded=convert_numbers(zomato1)
```

```python
zomato1_encoded
```

```python
corr=zomato1_encoded.corr()
plt.figure(figsize=(14,8))
sns.heatmap(corr,annot=True,cmap='Set2')
```

```python
ax=sns.countplot(x=zomato11_copy['online_order'])
for bars in ax.containers:
    ax.bar_label(bars)
fig=plt.gcf()
fig.set_size_inches(5,3)
plt.title("Restaurants Delivering Online or Not")
plt.show()
```

```python
ax=sns.countplot(x=zomato11_copy['online_order'],palette='rainbow')
for bars in ax.containers:
    ax.bar_label(bars)
fig=plt.gcf()
fig.set_size_inches(5,3)
plt.title("Restaurants Delivering Online or Not")
plt.show()
```

```python
ax=sns.countplot(x=zomato11_copy['book_table'],palette='dark')
for bars in ax.containers:
    ax.bar_label(bars)
fig=plt.gcf()
fig.set_size_inches(5,3)
plt.title("Restaurants Providing Table Booking or Not")
plt.show()
```

```python
c=pd.crosstab(zomato11_copy['rate'],zomato11_copy['book_table'])
```

```python
c
```

```python
c.sum(1)
```

```python
c.sum(1).astype(float)
```

```python
c.div(c.sum(1).astype(float),axis=0)
```

```python
c.div(c.sum(1).astype(float),axis=0).plot(kind='bar',stacked=True,color=['blue','red'])
plt.title("Table Booking Vs Normal Rest for Ratings")
plt.legend(loc=1)
plt.show()
```
